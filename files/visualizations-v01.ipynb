```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Load the preprocessed dataset
df = pd.read_parquet('../datasets/_dataset.parquet')  # Adjust path if needed

# 5. Top 10 Cities by Arrivals (Consistent Color Scheme)
top_arrival_cities = df.groupby('DestinationCity')['PassengerCount'].sum().reset_index()
top_arrival_cities = top_arrival_cities.sort_values(by='PassengerCount', ascending=True).tail(10)
fig5 = px.bar(top_arrival_cities, x='PassengerCount', y='DestinationCity', orientation='h', title="Top 10 Busiest Cities by Arrivals (Passenger Count)",
              color='PassengerCount', color_continuous_scale='Blues')  # Consistent color scale
fig5.update_layout(title_font_size=20, xaxis_title="Passenger Count", yaxis_title="City")
fig5.update_traces(texttemplate='%{x:.0f}', textposition='outside')
fig5.update_layout(xaxis=dict(autorange=True), yaxis=dict(autorange=True))
fig5.show()

# 6. Top 10 Cities by Departures (Consistent Color Scheme)
top_departure_cities = df.groupby('OriginCity')['PassengerCount'].sum().reset_index()
top_departure_cities = top_departure_cities.sort_values(by='PassengerCount', ascending=True).tail(10)
fig6 = px.bar(top_departure_cities, x='PassengerCount', y='OriginCity', orientation='h', title="Top 10 Busiest Cities by Departures (Passenger Count)",
              color='PassengerCount', color_continuous_scale='Oranges')
fig6.update_layout(title_font_size=20, xaxis_title="Passenger Count", yaxis_title="City")
fig6.update_traces(texttemplate='%{x:.0f}', textposition='outside')
fig6.update_layout(xaxis=dict(autorange=True), yaxis=dict(autorange=True))
fig6.show()

# 7. Top 10 Airports by Arrivals (Consistent Color Scheme)
top_arrival_airports = df.groupby('DestinationAirportCode')['PassengerCount'].sum().reset_index()
top_arrival_airports = top_arrival_airports.sort_values(by='PassengerCount', ascending=True).tail(10)
fig7 = px.bar(top_arrival_airports, x='PassengerCount', y='DestinationAirportCode', orientation='h', title="Top 10 Busiest Airports by Arrivals (Passenger Count)",
              color='PassengerCount', color_continuous_scale='Blues')
fig7.update_layout(title_font_size=20, xaxis_title="Passenger Count", yaxis_title="Airport Code")
fig7.update_traces(texttemplate='%{x:.0f}', textposition='outside')
fig7.update_layout(xaxis=dict(autorange=True), yaxis=dict(autorange=True))
fig7.show()

# 8. Top 10 Airports by Departures (Consistent Color Scheme)
top_departure_airports = df.groupby('OriginAirportCode')['PassengerCount'].sum().reset_index()
top_departure_airports = top_departure_airports.sort_values(by='PassengerCount', ascending=True).tail(10)
fig8 = px.bar(top_departure_airports, x='PassengerCount', y='OriginAirportCode', orientation='h', title="Top 10 Busiest Airports by Departures (Passenger Count)",
              color='PassengerCount', color_continuous_scale='Oranges')
fig8.update_layout(title_font_size=20, xaxis_title="Passenger Count", yaxis_title="Airport Code")
fig8.update_traces(texttemplate='%{x:.0f}', textposition='outside')
fig8.update_layout(xaxis=dict(autorange=True), yaxis=dict(autorange=True))
fig8.show()

# 9. Top 10 Routes by Passenger Count (Consistent Color Scheme)
top_routes = df.groupby('Route')['PassengerCount'].sum().reset_index()
top_routes = top_routes.sort_values(by='PassengerCount', ascending=True).tail(10)
fig9 = px.bar(top_routes, x='PassengerCount', y='Route', orientation='h', title="Top 10 Busiest Routes by Passenger Count",
              color='PassengerCount', color_continuous_scale='Blues')
fig9.update_layout(title_font_size=20, xaxis_title="Passenger Count", yaxis_title="Route")
fig9.update_traces(texttemplate='%{x:.0f}', textposition='outside')
fig9.update_layout(xaxis=dict(autorange=True), yaxis=dict(autorange=True))
fig9.show()

# 10. Top 10 Longest Unique Routes (Horizontal Bar Chart)
unique_routes_df = df.drop_duplicates(subset=['Route'])
top_longest_routes = unique_routes_df.sort_values(by='RouteDistanceInMiles', ascending=True).tail(10)
fig10 = px.bar(top_longest_routes, x='RouteDistanceInMiles', y='Route', orientation='h', title="Top 10 Longest Routes by Distance (Miles)")
fig10.update_traces(marker_color='teal')
fig10.update_layout(title_font_size=20, xaxis_title="Route Distance (Miles)", yaxis_title="Route")
fig10.update_layout(xaxis=dict(autorange=True), yaxis=dict(autorange=True))
fig10.show()

# 11. Top 10 Shortest Unique Routes (Horizontal Bar Chart)
top_shortest_routes = unique_routes_df.sort_values(by='RouteDistanceInMiles').head(10)
fig11 = px.bar(top_shortest_routes, x='RouteDistanceInMiles', y='Route', orientation='h', title="Top 10 Shortest Routes by Distance (Miles)")
fig11.update_traces(marker_color='salmon')
fig11.update_layout(title_font_size=20, xaxis_title="Route Distance (Miles)", yaxis_title="Route")
fig11.update_layout(xaxis=dict(autorange=True), yaxis=dict(autorange=True))
fig11.show()
```

```python
# 4. Passenger Count by Year (Line Chart in Gray Scale with Average Line)
passenger_count_by_year = df.groupby('Year')['PassengerCount'].sum().reset_index()
average_count = passenger_count_by_year['PassengerCount'].mean()

fig4 = px.line(passenger_count_by_year, x='Year', y='PassengerCount', title="Passengers Trend Over Years")
fig4.update_traces(mode='lines+markers', line=dict(color='gray'))
fig4.add_shape(type="line", x0=passenger_count_by_year['Year'].min(), x1=passenger_count_by_year['Year'].max(),
               y0=average_count, y1=average_count, line=dict(dash="dash", color="red"))  # Average line
fig4.add_annotation(x=passenger_count_by_year['Year'].max(), y=average_count, text=f"Avg: {average_count:.0f}", showarrow=False)
fig4.update_layout(title_font_size=20, xaxis_title="Year", yaxis_title="Passenger Count (Log Scale)", yaxis_type="log")
fig4.update_layout(xaxis=dict(autorange=True), yaxis=dict(autorange=True))
fig4.show()
```

```python
import plotly.express as px
import pandas as pd

# Define categories based on route distance
df['DistanceCategory'] = pd.cut(df['RouteDistanceInMiles'], bins=[0, 500, 1500, 3000], labels=['Short', 'Medium', 'Long'])

# Group by Year and Distance Category to sum up PassengerCount
if 'Year' in df.columns:
    trend_data = df.groupby(['Year', 'DistanceCategory'])['PassengerCount'].sum().reset_index()

# Define color palette for Distance Categories
distance_colors = {
    "Short": "dodgerblue",
    "Medium": "orange",
    "Long": "green"
}

# 1. Line plot for trends over time by Distance Category with markers
fig = px.line(trend_data, x='Year', y='PassengerCount', color='DistanceCategory', 
              title="Passenger Count Trends by Route Distance Category Over Time",
              line_shape='spline', markers=True, color_discrete_map=distance_colors)

# Add peak annotations for each Distance Category
for category in trend_data['DistanceCategory'].unique():
    category_data = trend_data[trend_data['DistanceCategory'] == category]
    peak_point = category_data.loc[category_data['PassengerCount'].idxmax()]
    fig.add_annotation(x=peak_point['Year'], y=peak_point['PassengerCount'],
                       text=f"Peak ({category})", showarrow=True, arrowhead=2, yshift=10,
                       font=dict(color=distance_colors[category]))

# Update layout for improved readability
fig.update_layout(
    xaxis_title="Year",
    yaxis_title="Total Passenger Count",
    legend_title="Distance Category",
    hovermode="x unified",
    title_x=0.5
)
fig.update_layout(xaxis=dict(autorange=True), yaxis=dict(autorange=True))
fig.show()
```

```python
import plotly.express as px
import plotly.graph_objects as go

# 1. Improved Line Plot for Passenger Count Over Time
passenger_trend = df.groupby('Year')['PassengerCount'].sum().reset_index()

# Create line plot with markers
fig1 = px.line(passenger_trend, x='Year', y='PassengerCount', title="Yearly Trend of Passenger Count",
               line_shape='spline', markers=True)

# Highlight peak and lowest points with annotations
peak_year = passenger_trend.loc[passenger_trend['PassengerCount'].idxmax()]
low_year = passenger_trend.loc[passenger_trend['PassengerCount'].idxmin()]
fig1.add_annotation(x=peak_year['Year'], y=peak_year['PassengerCount'],
                    text="Peak", showarrow=True, arrowhead=2)
fig1.add_annotation(x=low_year['Year'], y=low_year['PassengerCount'],
                    text="Lowest", showarrow=True, arrowhead=2)

# Update layout
fig1.update_layout(
    yaxis_title="Total Passenger Count",
    xaxis_title="Year",
    legend_title="Metric",
    hovermode="x unified",
    title_x=0.5
)
fig1.update_layout(xaxis=dict(autorange=True), yaxis=dict(autorange=True))
fig1.show()

# 2. Improved Line Plot for Average Fare Over Time
fare_trend = df.groupby('Year')['AverageFare'].mean().reset_index()

# Create line plot with markers
fig2 = px.line(fare_trend, x='Year', y='AverageFare', title="Yearly Trend of Average Fare",
               line_shape='spline', markers=True, color_discrete_sequence=['indianred'])

# Highlight peak and lowest points with annotations
peak_fare = fare_trend.loc[fare_trend['AverageFare'].idxmax()]
low_fare = fare_trend.loc[fare_trend['AverageFare'].idxmin()]
fig2.add_annotation(x=peak_fare['Year'], y=peak_fare['AverageFare'],
                    text="Highest Fare", showarrow=True, arrowhead=2)
fig2.add_annotation(x=low_fare['Year'], y=low_fare['AverageFare'],
                    text="Lowest Fare", showarrow=True, arrowhead=2)

# Update layout
fig2.update_layout(
    yaxis_title="Average Fare",
    xaxis_title="Year",
    legend_title="Metric",
    hovermode="x unified",
    title_x=0.5
)
fig2.update_layout(xaxis=dict(autorange=True), yaxis=dict(autorange=True))
fig2.show()
```

```python
import plotly.express as px
import pandas as pd

# Aggregate data by Year and Quarter
quarterly_trends = df.groupby(['Year', 'Quarter']).agg({'PassengerCount': 'sum', 'AverageFare': 'mean'}).reset_index()

# Convert Quarter to string format to match color dictionary keys (e.g., 1 -> "Q1")
quarterly_trends['Quarter'] = "Q" + quarterly_trends['Quarter'].astype(str)

# Define a consistent color palette for the quarters
quarter_colors = {
    "Q1": "royalblue",
    "Q2": "orange",
    "Q3": "green",
    "Q4": "red"
}

# 1. Line Plot for Passenger Count Trends by Quarter Over Time
fig1 = px.line(quarterly_trends, x='Year', y='PassengerCount', color='Quarter',
               title="Passenger Count Trends by Quarter Over Time",
               color_discrete_map=quarter_colors)

# Add markers for each data point
fig1.update_traces(mode="lines+markers")

# Customize layout
fig1.update_layout(
    xaxis_title="Year",
    yaxis_title="Passenger Count",
    legend_title="Quarter",
    hovermode="x unified"
)
fig1.update_layout(xaxis=dict(autorange=True), yaxis=dict(autorange=True))
fig1.show()

# 2. Line Plot for Average Fare Trends by Quarter Over Time
fig2 = px.line(quarterly_trends, x='Year', y='AverageFare', color='Quarter',
               title="Average Fare Trends by Quarter Over Time",
               color_discrete_map=quarter_colors)

# Add markers for each data point
fig2.update_traces(mode="lines+markers")

# Customize layout
fig2.update_layout(
    xaxis_title="Year",
    yaxis_title="Average Fare",
    legend_title="Quarter",
    hovermode="x unified"
)
fig2.update_layout(xaxis=dict(autorange=True), yaxis=dict(autorange=True))
fig2.show()
```
