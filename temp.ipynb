```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Load the preprocessed dataset
df = pd.read_parquet('datasets/_dataset.parquet')  # Adjust path if needed

# 1. Column Statistics - Total and Unique Counts (Indicators)
columns_to_check = [
    'Year', 'Quarter', 'OriginCityMarketId', 'DestinationCityMarketId',
    'OriginCity', 'DestinationCity', 'OriginAirportId', 'DestinationAirportId', 
    'OriginAirportCode', 'DestinationAirportCode', 'RouteDistanceInMiles', 'PassengerCount', 
    'AverageFare', 'LargestCarrierCode', 'LargestCarrierMarketShare', 'LargestCarrierAverageFare', 
    'LowestFareCarrierCode', 'LowestFareMarketShare', 'LowestFare', 
    'OriginCityCoordinates', 'DestinationCityCoordinates', 'Route'
]

# Calculate total and unique counts
table_data = []
for column in columns_to_check:
    if column in df.columns:
        total_count = df[column].count()
        unique_count = df[column].nunique()
        table_data.append({'Column': column, 'Total Count': total_count, 'Unique Count': unique_count})

# Display indicators for total and unique counts
for data in table_data:
    #fig = go.Figure()
    #fig.add_trace(go.Indicator(
    #    mode="number",
    #    value=data['Total Count'],
    #    title={'text': f"Total Count - {data['Column']}"}))
    #fig.show()

    fig = go.Figure()
    fig.add_trace(go.Indicator(
        mode="number",
        value=data['Unique Count'],
        title={'text': f"Unique Count - {data['Column']}"}))
    fig.show()

# 2. Unique Aggregations (Unique Counts for Combined Columns - Indicators)
unique_counts = {
    'Unique Cities': len(pd.concat([df['OriginCity'], df['DestinationCity']]).unique()),
    'Unique Airports': len(pd.concat([df['OriginAirportCode'], df['DestinationAirportCode']]).unique()),
    'Unique City Market IDs': len(pd.concat([df['OriginCityMarketId'], df['DestinationCityMarketId']]).unique()),
    'Unique Airport IDs': len(pd.concat([df['OriginAirportId'], df['DestinationAirportId']]).unique()),
    'Unique Carrier Codes': len(pd.concat([df['LargestCarrierCode'], df['LowestFareCarrierCode']]).unique()),
    'Unique Routes': df['Route'].nunique()
}

# Display unique count indicators for combined columns
for key, value in unique_counts.items():
    fig = go.Figure(go.Indicator(
        mode="number",
        value=value,
        title={'text': key}))
    fig.show()

# 3. Total Passenger Count (Single Value Display)
total_passenger_count = df['PassengerCount'].sum()
fig3 = go.Figure(go.Indicator(mode="number", value=total_passenger_count, title="Total Passenger Count"))
fig3.show()

# 4. Passenger Count by Year (Line Chart)
passenger_count_by_year = df.groupby('Year')['PassengerCount'].sum().reset_index()
fig4 = px.line(passenger_count_by_year, x='Year', y='PassengerCount', title="Passenger Count by Year")
fig4.show()

# 5. Top 10 Cities by Arrivals (Bar Chart)
top_arrival_cities = df.groupby('DestinationCity')['PassengerCount'].sum().reset_index()
top_arrival_cities = top_arrival_cities.sort_values(by='PassengerCount', ascending=False).head(10)
fig5 = px.bar(top_arrival_cities, x='DestinationCity', y='PassengerCount', title="Top 10 Cities by Arrivals (Passenger Count)")
fig5.show()

# 6. Top 10 Cities by Departures (Bar Chart)
top_departure_cities = df.groupby('OriginCity')['PassengerCount'].sum().reset_index()
top_departure_cities = top_departure_cities.sort_values(by='PassengerCount', ascending=False).head(10)
fig6 = px.bar(top_departure_cities, x='OriginCity', y='PassengerCount', title="Top 10 Cities by Departures (Passenger Count)")
fig6.show()

# 7. Top 10 Airports by Arrivals (Bar Chart)
top_arrival_airports = df.groupby('DestinationAirportCode')['PassengerCount'].sum().reset_index()
top_arrival_airports = top_arrival_airports.sort_values(by='PassengerCount', ascending=False).head(10)
fig7 = px.bar(top_arrival_airports, x='DestinationAirportCode', y='PassengerCount', title="Top 10 Airports by Arrivals (Passenger Count)")
fig7.show()

# 8. Top 10 Airports by Departures (Bar Chart)
top_departure_airports = df.groupby('OriginAirportCode')['PassengerCount'].sum().reset_index()
top_departure_airports = top_departure_airports.sort_values(by='PassengerCount', ascending=False).head(10)
fig8 = px.bar(top_departure_airports, x='OriginAirportCode', y='PassengerCount', title="Top 10 Airports by Departures (Passenger Count)")
fig8.show()

# 9. Top 10 Routes by Passenger Count (Bar Chart)
top_routes = df.groupby('Route')['PassengerCount'].sum().reset_index()
top_routes = top_routes.sort_values(by='PassengerCount', ascending=False).head(10)
fig9 = px.bar(top_routes, x='Route', y='PassengerCount', title="Top 10 Routes by Passenger Count")
fig9.show()

# 10. Top 10 Longest Unique Routes (Bar Chart)
unique_routes_df = df.drop_duplicates(subset=['Route'])
top_longest_routes = unique_routes_df.sort_values(by='RouteDistanceInMiles', ascending=False).head(10)
fig10 = px.bar(top_longest_routes, x='Route', y='RouteDistanceInMiles', title="Top 10 Longest Routes by Distance (Miles)")
fig10.show()

# 11. Top 10 Shortest Unique Routes (Bar Chart)
top_shortest_routes = unique_routes_df.sort_values(by='RouteDistanceInMiles').head(10)
fig11 = px.bar(top_shortest_routes, x='Route', y='RouteDistanceInMiles', title="Top 10 Shortest Routes by Distance (Miles)")
fig11.show()
```
