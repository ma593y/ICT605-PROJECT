```python
import pandas as pd
import plotly.graph_objects as go

# Load the preprocessed dataset
df = pd.read_parquet('datasets/_dataset.parquet')  # Adjust path if needed

# Unique Year Count
unique_count_year = df['Year'].nunique()
go.Figure(go.Indicator(mode="number", value=unique_count_year, title={'text': "Unique Year Count"})).show()

# Unique Quarter Count
unique_count_quarter = df['Quarter'].nunique()
go.Figure(go.Indicator(mode="number", value=unique_count_quarter, title={'text': "Unique Quarter Count"})).show()

# Unique Cities (combined OriginCity and DestinationCity)
unique_cities_count = len(pd.concat([df['OriginCity'], df['DestinationCity']]).unique())
go.Figure(go.Indicator(mode="number", value=unique_cities_count, title={'text': "Unique Cities (combined OriginCity and DestinationCity)"})).show()

# Unique OriginCity Count
unique_count_origin_city = df['OriginCity'].nunique()
go.Figure(go.Indicator(mode="number", value=unique_count_origin_city, title={'text': "Unique OriginCity Count"})).show()

# Unique DestinationCity Count
unique_count_dest_city = df['DestinationCity'].nunique()
go.Figure(go.Indicator(mode="number", value=unique_count_dest_city, title={'text': "Unique DestinationCity Count"})).show()

# Unique Airports (combined OriginAirportCode and DestinationAirportCode)
unique_airports_count = len(pd.concat([df['OriginAirportCode'], df['DestinationAirportCode']]).unique())
go.Figure(go.Indicator(mode="number", value=unique_airports_count, title={'text': "Unique Airports (combined OriginAirportCode and DestinationAirportCode)"})).show()

# Unique OriginAirportCode Count
unique_count_origin_airport_code = df['OriginAirportCode'].nunique()
go.Figure(go.Indicator(mode="number", value=unique_count_origin_airport_code, title={'text': "Unique OriginAirportCode Count"})).show()

# Unique DestinationAirportCode Count
unique_count_dest_airport_code = df['DestinationAirportCode'].nunique()
go.Figure(go.Indicator(mode="number", value=unique_count_dest_airport_code, title={'text': "Unique DestinationAirportCode Count"})).show()

# Unique Carrier Codes (combined LargestCarrierCode and LowestFareCarrierCode)
unique_carrier_codes_count = len(pd.concat([df['LargestCarrierCode'], df['LowestFareCarrierCode']]).unique())
go.Figure(go.Indicator(mode="number", value=unique_carrier_codes_count, title={'text': "Unique Carrier Codes (combined LargestCarrierCode and LowestFareCarrierCode)"})).show()

# Unique LargestCarrierCode Count
unique_count_largest_carrier = df['LargestCarrierCode'].nunique()
go.Figure(go.Indicator(mode="number", value=unique_count_largest_carrier, title={'text': "Unique LargestCarrierCode Count"})).show()

# Unique LowestFareCarrierCode Count
unique_count_lowest_fare_carrier = df['LowestFareCarrierCode'].nunique()
go.Figure(go.Indicator(mode="number", value=unique_count_lowest_fare_carrier, title={'text': "Unique LowestFareCarrierCode Count"})).show()

# Unique Routes Count
unique_routes_count = df['Route'].nunique()
go.Figure(go.Indicator(mode="number", value=unique_routes_count, title={'text': "Unique Routes Count"})).show()
```
